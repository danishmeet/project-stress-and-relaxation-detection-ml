```python
import pandas as pd

df = pd.read_csv("../data/processed/windows_features.csv")

df["Data_Source"] = "Real"

df.head()
```

```python
df["Stress_State"] = df["Label"].apply(
    lambda x: 1 if x == 2 else 0
)
```

```python
df["HRV_HR_Ratio"] = df["RMSSD"] / df["Mean_HR"]
```

```python
df["Resp_Regularity"] = 1 / (df["Resp_Variability"] + 1e-6)
```

```python
df["Autonomic_Index"] = (
    df["RMSSD"] +
    df["SDNN"] -
    df["Mean_HR"]
)
```

```python
df.to_csv("../data/processed/windows_features_engineered.csv", index=False)
```
