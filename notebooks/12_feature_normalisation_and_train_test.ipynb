```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/windows_features_engineered.csv")
df.shape
```

```python
# Stress = 1, Non-Stress (Baseline + Relaxed) = 0
y = df["Stress_State"]
```

```python
FEATURE_COLS = [
    "Mean_RR",
    "SDNN",
    "RMSSD",
    "Mean_HR",
    "Resp_Rate",
    "Resp_Variability",
    "HRV_HR_Ratio",
    "Resp_Regularity",
    "Autonomic_Index"
]

X = df[FEATURE_COLS]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.3,
    random_state=42,
    stratify=y
)
```

```python
print("Train size:", X_train.shape)
print("Test size:", X_test.shape)

print("\nTrain label distribution:")
print(y_train.value_counts(normalize=True))

print("\nTest label distribution:")
print(y_test.value_counts(normalize=True))
```

```python
np.save("../data/processed/X_train_eng.npy", X_train)
np.save("../data/processed/X_test_eng.npy", X_test)
np.save("../data/processed/y_train_eng.npy", y_train)
np.save("../data/processed/y_test_eng.npy", y_test)
```
